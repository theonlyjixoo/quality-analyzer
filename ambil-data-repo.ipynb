IMPORT LIBRARY

```python
import requests
import pandas as pd
#import matplotlib.pyplot as plt
# import seaborn as sns
```

TARIK DATA DARI GIT

```python

def fetch_github_data(topic, language, per_page=100):
    base_url = 'https://api.github.com/search/repositories'
    page = 1
    all_results = []

    while True:
        params = {
            'q': f'topic:{topic} language:{language}',
            'per_page': per_page,
            'page': page
        }

        response = requests.get(base_url, params=params)

        if response.status_code != 200:
            print(f"Error: {response.status_code}")
            break

        data = response.json()
        items = data.get('items', [])

        if not items:
            # No more data to fetch
            break

        all_results.extend(items)
        page += 1

    return all_results

# Contoh pemanggilan fungsi
topic_to_search = 'machine learning'
language_to_search = 'python'
result = fetch_github_data(topic_to_search, language_to_search)

#simpan menjadi dataframe pandas
df = pd.DataFrame(result)
df.to_csv('github_repositories.csv', index=False)

# Tampilkan hasil
for index, repo in enumerate(result, start=1):
    print(f"{index}. {repo['name']} - {repo['html_url']}")
```
