```python
def fetch_github_data(topic, language, per_page=100):
    base_url = 'https://api.github.com/search/repositories'
    page = 1
    all_results = []

    while True:
        params = {
            'q': f'topic:{topic} language:{language}',
            'per_page': per_page,
            'page': page
        }

        response = requests.get(base_url, params=params)

        if response.status_code != 200:
            print(f"Error: {response.status_code}")
            break

        data = response.json()
        items = data.get('items', [])

        if not items:
            # No more data to fetch
            break

        all_results.extend(items)
        page += 1

    return all_results

# Contoh pemanggilan fungsi
topic_to_search = 'machine learning'
language_to_search = 'python'
result = fetch_github_data(topic_to_search, language_to_search)

# Tampilkan hasil
for index, repo in enumerate(result, start=1):
    print(f"{index}. {repo['name']} - {repo['html_url']}")
```

```python
def get_repository_info(owner, repo):
    base_url = f'https://api.github.com/repos/{owner}/{repo}'
    response = requests.get(base_url)

    if response.status_code == 200:
        repo_info = response.json()
        return repo_info
    else:
        print(f"Error: {response.status_code}")
        return None
```

```python
def download_files_from_github(csv_file, token=None):
    # Read CSV file to get GitHub repository URLs
    with open(csv_file, newline='') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            github_url = row[0]  # Assuming GitHub repository URL is in the first column

            # Get the repository name from the URL
            parts = github_url.split('/')
            owner, repo_name = parts[-2], parts[-1]

            # Prepare the API endpoint to list repository contents
            api_url = f"https://api.github.com/repos/{owner}/{repo_name}/contents"

            # Send GET request to GitHub API to list repository contents
            headers = {'Accept': 'application/vnd.github.v3+json'}
            if token:
                headers['Authorization'] = f'token {token}'

            response = requests.get(api_url, headers=headers)

            if response.status_code == 200:
                contents = response.json()

                # Create a folder for the repository
                folder_name = f"{owner}_{repo_name}"
                if not os.path.exists(folder_name):
                    os.makedirs(folder_name)

                for item in contents:
                    if item['type'] == 'file' and item['name'].endswith('.py'):
                        # Prepare the download URL for Python files
                        download_url = item['download_url']
                        
                        # Send GET request to download Python file
                        response = requests.get(download_url)
                        if response.status_code == 200:
                            # Save the content to a file inside the repository folder
                            file_path = os.path.join(folder_name, item['name'])
                            with open(file_path, 'wb') as f:
                                f.write(response.content)
                            print(f"File {item['name']} downloaded successfully from {github_url}")
                        else:
                            print(f"Failed to download file from {download_url}. Status code: {response.status_code}")
            else:
                print(f"Failed to access repository {github_url}. Status code: {response.status_code}")

if __name__ == "__main__":
    csv_file = 'url_git.csv'  # path file url
    token = '####'  # personal git token
    download_files_from_github(csv_file, token)
```

```python
import os
import csv
import subprocess
from collections import defaultdict

def collect_python_files(root_dir):
    """Mengumpulkan semua file .py dari root_dir dan subfoldernya."""
    python_files = defaultdict(list)
    for subdir, _, files in os.walk(root_dir):
        for file in files:
            if file.endswith('.py'):
                full_path = os.path.join(subdir, file)
                folder_name = os.path.basename(subdir)
                python_files[folder_name].append(full_path)
    return python_files

def analyze_code(file_path):
    """Melakukan analisis statik menggunakan pylint dan menghitung line of code."""
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
    except UnicodeDecodeError:
        # Jika terjadi error, coba dengan encoding yang berbeda
        with open(file_path, 'r', encoding='latin-1') as file:
            lines = file.readlines()
    
    line_count = len(lines)
    
    result = subprocess.run(['pylint', file_path, '--disable=all', '--enable=warning,error'],
                            capture_output=True, text=True, encoding='utf-8')
    
    pylint_output = result.stdout
    
    bad_smells = sum(1 for line in pylint_output.split('\n') if 'warning' in line or 'error' in line)
    
    return line_count, bad_smells

def analyze_folder(python_files):
    """Melakukan analisis untuk setiap file dalam folder."""
    results = []
    for folder, files in python_files.items():
        for file in files:
            line_count, bad_smells = analyze_code(file)
            results.append((folder, os.path.basename(file), line_count, bad_smells))
    return results

def save_to_csv(results, output_file):
    """Menyimpan hasil analisis ke dalam file CSV."""
    with open(output_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Folder', 'File Name', 'Line Count', 'Bad Smells'])
        writer.writerows(results)

if __name__ == '__main__':
    root_directory = 'D:/S2/Tesis/code/repo'  # Ganti dengan path folder root anda
    output_csv = 'analysis_results.csv'
    
    python_files = collect_python_files(root_directory)
    results = analyze_folder(python_files)
    save_to_csv(results, output_csv)
    
    print(f'Hasil analisis telah disimpan dalam {output_csv}')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Skalakan data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(final_df[['total_bad_smells', 'total_line_count']])

# Metode Elbow untuk menentukan jumlah cluster yang optimal
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

# Plot metode Elbow
plt.figure(figsize=(10, 6))
plt.plot(range(1, 10), inertia, marker='o', linestyle='--')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Inertia')
plt.title('Metode Elbow untuk Menentukan Jumlah Cluster')
plt.show()
```

```python
# Melakukan clustering dengan K-Means
k = 3  # Misalnya, pilih jumlah cluster yang optimal dari plot elbow
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

# Menambahkan label cluster ke dalam dataframe
final_df['cluster'] = clusters

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_line_count', y='total_bad_smells', data=final_df, hue='cluster', palette='viridis', legend='full')
plt.title('Clustering Total Bad Smells dengan Line of Code')
plt.xlabel('Total Line Count')
plt.ylabel('Total Bad Smells')
plt.show()
```

```python
# Memilih kolom yang relevan untuk clustering
clustering_data = final_df[['total_bad_smells', 'total_line_count']].dropna()

# Menentukan jumlah cluster (misalnya 3)
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(clustering_data)

# Menambahkan label cluster ke data
clustering_data['cluster'] = kmeans.labels_

# Visualisasi hasil clustering
plt.figure(figsize=(12, 6))
sns.scatterplot(x='total_bad_smells', y='total_line_count', hue='cluster', data=clustering_data, palette='viridis')
plt.title('Clustering antara Total Bad Smells dan Total Line Count')
plt.xlabel('Total Bad Smells')
plt.ylabel('Total Line Count')
plt.legend(title='Cluster')
plt.show()
```

```python
print("\nDeskripsi statistik setiap cluster:")
for cluster_num in range(num_clusters):
    cluster_data = clustering_data[clustering_data['cluster'] == cluster_num]
    print(f"\nCluster {cluster_num}:")
    print(cluster_data.describe())
    print(f"Jumlah proyek dalam cluster {cluster_num}: {len(cluster_data)}")
```

Dengan memahami distribusi dan hubungan antara stargazers dan watchers, pengembang dapat membuat keputusan yang lebih baik tentang bagaimana mengelola proyek dan komunitas mereka.
```python
# Konversi kolom waktu ke format datetime
final_df['updated_at'] = pd.to_datetime(final_df['updated_at'])
final_df['pushed_at'] = pd.to_datetime(final_df['pushed_at'])

# Menambahkan kolom tahun untuk agregasi
final_df['year'] = final_df['updated_at'].dt.year

# Mengelompokkan dan mengagregasi data berdasarkan tahun
yearly_data = final_df.groupby('year').agg({
    'stargazers_count': 'sum',
    'watchers_count': 'sum',
    'updated_at': 'count',  # Menghitung jumlah pembaruan dalam setahun
    'pushed_at': 'count'    # Menghitung jumlah push dalam setahun
}).reset_index()

# Visualisasi tren stargazers_count dan watchers_count seiring waktu
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='stargazers_count', data=yearly_data, label='Stargazers Count', marker='o')
sns.lineplot(x='year', y='watchers_count', data=yearly_data, label='Watchers Count', marker='o')
plt.title('Tren Perkembangan Stargazers dan Watchers')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.legend()
plt.grid(True)
plt.show()

# Visualisasi tren updated_at dan pushed_at seiring waktu
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='updated_at', data=yearly_data, label='Updated Count', marker='o')
sns.lineplot(x='year', y='pushed_at', data=yearly_data, label='Pushed Count', marker='o')
plt.title('Tren Aktivitas Pembaruan dan Push')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Aktivitas')
plt.legend()
plt.grid(True)
plt.show()
```
